## Regression Model Implementation

```python
import sklearn
from sklearn.tree import DecisionTreeRegressor as DTR
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.model_selection import train_test_split
```

```python
#from sklearn.preprocessing import LabelEncoder
#encoder=LabelEncoder()
```

```python
new_data
```

```python
feature_names = ['longitude', 'latitude','housing_median_age', 'total_rooms', 'total_bedrooms', 'ocean_distance', 'rooms_per_household']
```

```python
x = new_data[feature_names]
y = new_data['median_house_value']
```

```python
# I created a function that gives me the value for Mean Absolute Error (MAE) for my DTR model when i plug in my training and validation data
def get_mae(max_leaf_nodes, x_train, x_val, y_train, y_val):
  model = DTR(max_leaf_nodes = max_leaf_nodes, random_state=0)
  model.fit(x_train, y_train)
  val_prediction = model.predict(x_val)
  mae = MAE(y_val, val_prediction)
  return mae
```

```python
# Split my house dataset into my training and validation data

x_train, x_val, y_train, y_val = train_test_split(x, y, random_state=0)
```

```python
# This for loop helps my to check MAE usind different values for max_leaf_nodes. this will help me decide the best value for my max_leaf_node
# for the DTR model for my data

for max_leaf_nodes in [4, 40, 100, 200, 400, 600, 4000]:
  my_mae = get_mae(max_leaf_nodes, x_train, x_val, y_train, y_val)
  print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))

# choose the max_leaf_nodes with the lowest MAE which is 600
```

```python
# I define my model
home_model = DTR(max_leaf_nodes=600, random_state=1) #DTR - DecsionTreeRegressor model for the house prices
```

```python
# Fitting my model for my training data

home_model.fit(x_train, y_train)
```

```python
# Prediction for my validation data
validation_predictions = home_model.predict(x_val)
```

```python
print('Prediction for the house prices')
print(validation_predictions[1:10])
print('\n')
print('Actual Prices from the validation data')
print('\n')
print(y_val[1:10])
```

```python
print(MAE(y_val, validation_predictions)) #mean absolute error for my model
```

```python
print('Prediction for the house prices')
print(validation_predictions[0:5])
print('\n')
print('Actual Prices from the validation data')
print('\n')
print(y_val[0:5])
```

## Random Forest Regressor (RTR)

This is more sophiscated model approach. RTR uses many trees and makes prediction by averaging the prediction of each component tree. The predictive accuracy is much better than using a single decision tree.

```python
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.metrics import mean_absolute_error as MAE
```

```python
home_forest_model = RFR(random_state=1)
home_forest_model.fit(x_train, y_train)
forest_prediction = home_forest_model.predict(x_val)
print(MAE(y_val, forest_prediction))

# Random forest improve my model better than Decision Trees
```
